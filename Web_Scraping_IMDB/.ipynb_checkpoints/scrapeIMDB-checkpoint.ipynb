```python
import requests, openpyxl
from urllib.request import Request, urlopen
from bs4 import BeautifulSoup

url = "https://www.imdb.com/chart/top/"
```

```python
excel = openpyxl.Workbook()
sheet = excel.active
sheet.title = 'iMDB - Top Rated Movies'

sheet.append(['Movie Rank', 'Movie Name', 'Year of Release', 'Runtime', 'iMDB_rating'])
```

```python
req = Request(url , headers={'User-Agent': 'Mozilla/5.0'})
webpage = urlopen(req).read()                                 #Extract the HTML source code of the URL 
soup = BeautifulSoup(webpage, "html.parser")                      #Returns a beautiful soup object (Parsed HTML)
```

```python
movies = soup.find('ul', class_ = "ipc-metadata-list").find_all('li', class_ = "ipc-metadata-list-summary-item")
print(len(movies))
```

```python
try:
    for movie in movies:
        rank = movie.find('div', class_ = 'ipc-title').a.h3.text.split('.')[0]
        name = movie.find('div', class_ = 'ipc-title').a.h3.text.split('.')[1]
        year = movie.find('div', class_ = 'sc-c7e5f54-7').span.text
        runtime = movie.find('div', class_ = 'sc-c7e5f54-7').findChildren()[1].text        
        audience_rating = movie.find('span','sc-c7e5f54-1').find('div', class_ = 'sc-e3e7b191-0').span.text.split()[0]
        
        print(name, " ", rank, " ", year, " ", runtime, " ", audience_rating)
        sheet.append([rank, name, year, runtime, audience_rating])
        
except Exception as e:
    print(e)
```

```python
excel.save("iMDB_Top_Rated_Movies.xlsx")
```
